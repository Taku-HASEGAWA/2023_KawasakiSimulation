```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["text.usetex"] =True
plt.rcParams['font.family'] = 'Arial'
plt.rcParams["font.size"] = 25
from matplotlib import patheffects
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

fig, ax = plt.subplots(figsize=(6, 6))
nx = 101
x = np.linspace(0.0, np.pi/2, nx)
y = 0.375*np.cos(x)
ax.plot(x, y,linewidth = 1.5)
ax.tick_params(axis='x', labelsize=30)
ax.tick_params(axis='y', labelsize=30)
plt.tick_params(which='major',width = 1, length = 10)
plt.tick_params(which='minor',width = 1, length = 5)
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
plt.xticks(
    [0, np.pi/4, np.pi/2],
    [r"$0$", r"$\frac{\pi}{4}$", r"$\frac{\pi}{2}$"]
)
plt.yticks([0, 0.125*3, 0.125*4], [r"$0$", r"$\frac{l}{2}$", r"$\frac{d}{2}$"])
plt.fill_between(x,y,color='blue',alpha=0.5)
ax.set_xlim(0., np.pi/2)
ax.set_xlabel(r"$\theta$",fontsize=30)
ax.set_ylim(0., 0.5)
ax.set_ylabel(r"$y_{\mathrm{c}}$",fontsize=30)
plt.savefig('./buffon_1.png')
plt.savefig('./buffon_1.pdf')
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)
i, pi,error  = np.loadtxt("../../code/1_buffon/buffon-pi-error.dat", comments='#', unpack=True)
plt.plot(i, pi, "",linewidth = 1.25,markersize=10,color="b",label=r"rand()")
plt.xscale('log')
###Drawing a line ######
x= np.linspace(1e1, 5e14, 100) 
y= np.pi+0*x
plt.plot(x, y, "--",markersize=3,linewidth = 1.5, color="k",label=r"$\pi$")
#########
plt.tick_params(which='major',width = 1, length = 10)
plt.tick_params(which='minor',width = 1, length = 5)
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
ax.set_ylim(2.68, 3.22)
ax.set_xlim(1.e1, 1.e15)
plt.xlabel(r"$n$",color='k', size=30)
plt.ylabel(r"$\pi(n)$",color='k', size=30)
plt.minorticks_on()
plt.grid(axis='both')
plt.legend(ncol=1, loc=4, borderaxespad=0, fontsize=25,frameon=False)

plt.savefig('./buffon_3.png')
plt.savefig('./buffon_3.pdf')
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(111)
plt.plot(i, error, "",linewidth = 1.0,color="r",label=r"rand()")
plt.xscale('log')
plt.yscale('log')
###Drawing a line ######
x= np.linspace(1e1, 1e16, 100) 
y=3.25976064/x**0.5 #theoretical value from cycle of rand() 
plt.plot(x, y, "--",markersize=2,linewidth = 2.0, color="k",label=r"$C/\sqrt{n}$")
#########
plt.tick_params(which='major',width = 1, length = 10)
plt.tick_params(which='minor',width = 1, length = 5)
ax.spines['top'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)
ax.spines['left'].set_linewidth(3)
ax.spines['right'].set_linewidth(3)
ax.set_ylim(1.e-8, 1.e0)
ax.set_xlim(1.e1, 1.e15)
plt.xlabel(r"$n$",color='k', size=30)
plt.ylabel(r"$\delta(n)=\left|\pi-\pi(n)\right|$",color='k', size=30)
plt.minorticks_on()
plt.grid(axis='both')
plt.legend(ncol=1, loc=1, borderaxespad=0, fontsize=25,frameon=False)

plt.savefig('./buffon_4.png')
plt.savefig('./buffon_4.pdf')
```
